Exploring and Visualizing a Simple Dataset

1. Exploring dataset (Iris)

```python
# Importing Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading Iris DataSet

iris = sns.load_dataset('iris')
df = pd.DataFrame(iris)
print(df.head())
```

```python
# printing shape and columns of the dataset
print("Shape: ",df.shape)
print("Columns: ",df.columns)
```

```python
# Summary statistics
print(df.info())
print("\nDescription:\n ",df.describe())
```

2. Visualizing the dataset

```python
sns.scatterplot(data=df, x="sepal_length",y="sepal_width",hue="species")
plt.title("Relationship: Petal length vs Petal width")
plt.show()
```

```python
sns.histplot(data=df,x="petal_length",kde=True,hue="species")
plt.title("Distribution of Petal length")
plt.show()
```

```python
sns.boxplot(data=df,x="species",y="petal_width")
plt.title("Outlier detection for petal width")
plt.show()
```
